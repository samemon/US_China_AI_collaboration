#Plot productivity

```python
def plotPvalue(ax, x, y, pVal, ytop, lw=0.2, c='black'):
    """
    This function will be used to plot the p-values.
    """

    height = max(y[0], y[1])
    gap = ytop * 0.05
    vlength = ytop * 0.03

    yPos = height + gap + vlength

    ax.hlines(yPos, x[0], x[1], color=c, lw=lw)
    ax.vlines(x[0], height + gap, yPos, color=c, lw=lw)
    ax.vlines(x[1], height + gap, yPos, color=c, lw=lw)

    stars = '***' if pVal < 0.001 else '**' if pVal < 0.01 else '*' if pVal < 0.05 else f'p={round(pVal, 2)}'
    offset = 0.01 if pVal < 0.05 else 0.04

    if stars in ["***","**","*"]:
        ax.text((x[0] + x[1])/2, yPos + offset*ytop, stars, ha='center', color=c)
    else:
        ax.text((x[0] + x[1])/2, yPos + offset*ytop, stars, ha='center', color=c, fontsize=9)
```

```python
import numpy as np

# Team size aggregate as a box plot

china = pd.read_csv(path2+"CEM_china_us_data.csv")
us = pd.read_csv(path2+"CEM_us_china_data.csv")

# Old files with some anomalies which were fixed later by Bedoor post NHB
# china = pd.read_csv(path2+"CEM_china_us_data_bedoor.csv")
# us = pd.read_csv(path2+"CEM_us_china_data_bedoor.csv")


data = pd.concat([us,china])

us_color = '#779ECB'#'#829460'
china_color = '#FF6961'#'#E98EAD'
other_color = '#B6AFAC'#'#8B7E74'#'#4472c4'

plt.rcParams["figure.figsize"] = (8.3,3)

fig = plt.figure()

labelfontsize = 8
subtitlefontsize = 9
panellabelsize = 12

gs = fig.add_gridspec(1,3)
ax1 = fig.add_subplot(gs[:, :1])
ax2 = fig.add_subplot(gs[:, 1:2])
ax3 = fig.add_subplot(gs[:, 2:3])

####################
# PLOTTING PRODUCTIVITY
####################

data_prod = data[['PubYear','PID','Country','us_china_collab','numAuthors']].drop_duplicates()

# Removing papers with < 2 authors (post nHB review)
data_prod = data_prod[data_prod.numAuthors.gt(1)]

data_prod = data_prod.\
                groupby(['PubYear','Country','us_china_collab']).count().reset_index().rename(columns={'PID':'num_Papers'})

# Log scaling
data_prod['num_Papers'] = np.log10(data_prod['num_Papers'])

china_alone = data_prod[(data_prod.Country=="China") & (data_prod.us_china_collab==0) & data_prod.PubYear.le(2020)]
china_collab = data_prod[(data_prod.Country=="China") & (data_prod.us_china_collab==1) & data_prod.PubYear.le(2020)]
us_alone = data_prod[(data_prod.Country=="United States") & (data_prod.us_china_collab==0) & data_prod.PubYear.le(2020)]
us_collab = data_prod[(data_prod.Country=="United States") & (data_prod.us_china_collab==1) & data_prod.PubYear.le(2020)]
#ax1.set_yscale('log')
ax1.plot(china_alone.PubYear, china_alone.num_Papers, '-o',label=r'$(\mathit{China}, \neg\mathit{US})$', c=china_color,
        markeredgecolor=china_color, markerfacecolor='white')
ax1.plot(us_alone.PubYear, us_alone.num_Papers, '-o', label=r'$(\mathit{US}, \neg\mathit{China})$', c=us_color,
         markeredgecolor=us_color, markerfacecolor='white')
ax1.plot(china_collab.PubYear, china_collab.num_Papers, "--o", label=r'$(\mathit{China}, \mathit{US})$', c=china_color)
ax1.plot(us_collab.PubYear, us_collab.num_Papers, "--o", label=r'$(\mathit{US}, \mathit{China})$', c=us_color)
ax1.set_xlim(2004.5,2020.5)
#ax1.set_xticks(range(2005,2021), rotation=45)
ax1.tick_params(labelsize=labelfontsize)
ax1.set_xlabel('year', fontsize=subtitlefontsize)
ax1.set_ylabel('log(productivity)', fontsize=subtitlefontsize)

#ax1.legend(fontsize=labelfontsize, ncols=2, loc='lower center')
ax1.legend(fontsize=labelfontsize, ncols=2, bbox_to_anchor=(1.1, -0.25))


####################
# PLOTTING TEAM SIZE
####################

data_team = data[['PubYear','PID','Country','us_china_collab', 'numAuthors']].drop_duplicates()

# Removing the papers with team size == 1 (post NHB reivew)
data_team = data_team[data_team.numAuthors.gt(1)]

china_alone = data_team[(data_team.Country=="China") & (data_team.us_china_collab==0)]
china_collab = data_team[(data_team.Country=="China") & (data_team.us_china_collab==1)]
us_alone = data_team[(data_team.Country=="United States") & (data_team.us_china_collab==0)]
us_collab = data_team[(data_team.Country=="United States") & (data_team.us_china_collab==1)]

linewidth = 1

bp_us = dict(color=us_color, linestyle='-', linewidth=linewidth)
bp_china = dict(color=china_color, linestyle='-', linewidth=linewidth)

# median props
mp_us = dict(linestyle='-', linewidth=linewidth, color=us_color)
mp_china = dict(linestyle='-', linewidth=linewidth, color=china_color)

# mean props

mep_us = dict(marker='s', markerfacecolor=us_color, markeredgecolor=us_color)
mep_china = dict(marker='s', markerfacecolor=china_color, markeredgecolor=china_color)

ax2.boxplot(us_collab.numAuthors, positions=[1], labels=[r'$(\mathit{US}, \mathit{China})$'],
            boxprops=bp_us, medianprops=mp_us, showfliers=False, showmeans=True,
            meanprops=mep_us,widths=0.5)

ax2.boxplot(us_alone.numAuthors, positions=[2], labels=[r'$(\mathit{US}, \neg\mathit{China})$'],
            boxprops=bp_us, medianprops=mp_us, showfliers=False, showmeans=True,
            meanprops=mep_us,widths=0.5)

ax2.boxplot(china_collab.numAuthors, positions=[4], labels=[r'$(\mathit{China}, \mathit{US})$'],
            boxprops=bp_china, medianprops=mp_china, showfliers=False, showmeans=True,
            meanprops=mep_china, widths=0.5)

ax2.boxplot(china_alone.numAuthors, positions=[5], labels=[r'$(\mathit{China}, \neg\mathit{US})$'],
            boxprops=bp_china, medianprops=mp_china, showfliers=False, showmeans=True,
            meanprops=mep_china,widths=0.5)

ax2.set_ylabel('team size', fontsize=subtitlefontsize)
ax2.set_xlabel('collaboration type', fontsize=subtitlefontsize)

ax2.tick_params(labelsize=labelfontsize)
ax2.tick_params(axis='x', rotation=90)
ax2.set_ylim(bottom=0)
ax2.set_yticks([0, 2, 4, 6, 8, 10])

from scipy.stats import ttest_ind

# perform a t-test assuming unequal variances between the two groups

tchina, pchina = ttest_ind(china_collab.numAuthors, china_alone.numAuthors, equal_var=False)
tus, pus = ttest_ind(us_collab.numAuthors, us_alone.numAuthors, equal_var=False)

print(tchina, pchina, tus, pus)

plotPvalue(ax2, [1,2], [9,8], pus, 9, lw=0.2, c='black')
plotPvalue(ax2, [4,5], [9,8], pchina, 9, lw=0.2, c='black')

####################
# PLOTTING AFFILIATIONS
####################

top100_affids_w_c2 = df_papers.groupby(['AffID','Country','AffName'])['c_2'].sum().sort_values(ascending=False).reset_index().head(100)

data_aff = data[['PID','Country','us_china_collab','AffID','c_2', 'numAuthors']].drop_duplicates()

# Removing the papers with team size == 1 (post NHB reivew)
data_aff = data_aff[data_aff.numAuthors.gt(1)]

china_alone = data_aff[(data_aff.Country=="China") & (data_aff.us_china_collab==0)]
china_collab = data_aff[(data_aff.Country=="China") & (data_aff.us_china_collab==1)]
us_alone = data_aff[(data_aff.Country=="United States") & (data_aff.us_china_collab==0)]
us_collab = data_aff[(data_aff.Country=="United States") & (data_aff.us_china_collab==1)]

us_alone_top100_perc = us_alone[us_alone.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique()/us_alone.PID.nunique()
us_collab_top100_perc = us_collab[us_collab.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique()/us_collab.PID.nunique()

china_alone_top100_perc = china_alone[china_alone.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique()/china_alone.PID.nunique()
china_collab_top100_perc = china_collab[china_collab.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique()/china_collab.PID.nunique()

# Computing confidence intervals

import statsmodels.stats.proportion as prop
import numpy as np

def compute_ci(trials, successes):

    # calculate the proportion and the standard error
    prop_val = successes / trials
    se = np.sqrt(prop_val * (1 - prop_val) / trials)

    # calculate the confidence interval
    ci_low, ci_high = prop.proportion_confint(count=successes, nobs=trials, alpha=0.05, method='normal')

    return ci_low, ci_high

us_alone_ci95 = compute_ci(us_alone.PID.nunique(), us_alone[us_alone.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique())
us_collab_ci95 = compute_ci(us_collab.PID.nunique(), us_collab[us_collab.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique())

china_alone_ci95 = compute_ci(china_alone.PID.nunique(), china_alone[china_alone.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique())
china_collab_ci95 = compute_ci(china_collab.PID.nunique(), china_collab[china_collab.AffID.isin(top100_affids_w_c2.AffID.unique())].PID.nunique())


# Conducting significance test
import scipy.stats as stats

# For US
n1_us = us_alone.PID.nunique()   # number of trials for distribution 1
p1_us = us_alone_top100_perc   # probability of success for distribution 1
n2_us = us_collab.PID.nunique()   # number of trials for distribution 2
p2_us =  us_collab_top100_perc  # probability of success for distribution 2

# calculate the number of successes for each distribution
k1_us = int(n1_us * p1_us)
k2_us = int(n2_us * p2_us)

# perform the significance test using Fisher's exact test
_, pval_us = stats.fisher_exact([[k1_us, n1_us-k1_us], [k2_us, n2_us-k2_us]])

# For China

n1_ch = china_alone.PID.nunique()   # number of trials for distribution 1
p1_ch = china_alone_top100_perc   # probability of success for distribution 1
n2_ch = china_collab.PID.nunique()   # number of trials for distribution 2
p2_ch =  china_collab_top100_perc  # probability of success for distribution 2


# calculate the number of successes for each distribution
k1_ch = int(n1_ch * p1_ch)
k2_ch = int(n2_ch * p2_ch)

# perform the significance test using Fisher's exact test
_, pval_ch = stats.fisher_exact([[k1_ch, n1_ch-k1_ch], [k2_ch, n2_ch-k2_ch]])

print("pvalues US, China:",pval_us, pval_ch)

# Plotting bars
width = 0.25
x = np.arange(2)  # the label locations

# Plotting collaborations
rects1 = ax3.bar(x - width/2, [p2_us*100, p2_ch*100], width, label='Treatment',
                color = [us_color, china_color], edgecolor = "black",
                 yerr=([us_collab_ci95[0],china_collab_ci95[0]],
                       [us_collab_ci95[1],china_collab_ci95[1]]))

# Plotting alone
rects2 = ax3.bar(x + width/2, [p1_us*100, p1_ch*100], width, label='Control',
                color = [us_color, china_color], hatch='/',edgecolor = "black",
                 yerr=([us_alone_ci95[0],china_alone_ci95[0]],
                       [us_alone_ci95[1],china_alone_ci95[1]]))

plotPvalue(ax3, [0-width/2 ,width/2], [45,40], pval_us, 45, lw=0.2, c='black')
plotPvalue(ax3, [1-width/2, 1+width/2], [37,25], pval_ch, 37, lw=0.2, c='black')

xticks = [0-width/2 ,width/2, 1-width/2, 1+width/2]


ax3.set_xticks(xticks)
ax3.set_xticklabels([r'$(\mathit{US}, \mathit{China})$', r'$(\mathit{US}, \neg\mathit{China})$', r'$(\mathit{China}, \mathit{US})$', r'$(\mathit{China}, \neg\mathit{US})$'])

ax3.tick_params(labelsize=labelfontsize)
ax3.tick_params(axis='x', rotation=90)

ax3.set_ylabel('percentage of last-author\naffiliations falling among\nthe 100 most impactful', fontsize=subtitlefontsize)
ax3.set_xlabel('collaboration type', fontsize=subtitlefontsize)

ax1.spines[['right', 'top']].set_visible(False)
ax2.spines[['right', 'top']].set_visible(False)
ax3.spines[['right', 'top']].set_visible(False)

plt.tight_layout()
plt.show()
```
